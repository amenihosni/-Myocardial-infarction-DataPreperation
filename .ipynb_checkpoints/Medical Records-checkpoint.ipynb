4.3 Handling Outliers


detecting outliers


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = 'MI.data'
df = pd.read_csv(file_path, header=None, na_values='?')

numeric_columns = df.select_dtypes(include='number').columns  

num_features = len(numeric_columns)
ncols = 3 
nrows = (num_features + ncols - 1) // ncols 


fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))
axs = axs.flatten()  # Aplatir la matrice d'axes

for i, column in enumerate(numeric_columns):
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1


    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifier les outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

   
    sns.histplot(df[column], bins=30, kde=False, color='lightblue', ax=axs[i])  

   
    axs[i].axvline(lower_bound, color='orange', linestyle='--', label='Lower Bound')
    axs[i].axvline(upper_bound, color='purple', linestyle='--', label='Upper Bound')

    
    if not outliers.empty:
        axs[i].scatter(outliers[column], [0.5] * len(outliers), color='red', edgecolor='black', s=100, label='Outliers')

    axs[i].set_title(f'Histogram of {column} with Thresholds and Outliers', fontsize=16)
    axs[i].set_xlabel(column, fontsize=14)
    axs[i].set_ylabel('Frequency', fontsize=14)
    axs[i].legend()


for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])


plt.tight_layout()
plt.show()
```

Handling outliers

Given the high percentage of outliers in several columns, the best strategy would be capping the outliers (Winsorization). This ensures that the structure of your dataset remains intact while limiting the influence of extreme values. Dropping the records would result in losing too much data, especially in columns with over 10% outliers

capping the outliers

```python
df_capped = df.copy()

lower_percentile = 0.05
upper_percentile = 0.95

for column in numeric_columns:
    lower_cap = df[column].quantile(lower_percentile)
    upper_cap = df[column].quantile(upper_percentile)
    
    # Cap the outliers
    df_capped[column] = df[column].clip(lower=lower_cap, upper=upper_cap)

print(df_capped.describe())
```

# 5. Feature selection


```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.impute import SimpleImputer
X = df_filled.iloc[:, :-12] 
y = df_filled.iloc[:, -12:] 


y_categorical = y.apply(lambda col: col.astype('category'))


imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

chi_selector = SelectKBest(chi2, k=10)
X_kbest = chi_selector.fit_transform(X_imputed, y_categorical.iloc[:, 0])  

selected_features = X.columns[chi_selector.get_support()]
print("Selected Features: ", selected_features)
```
